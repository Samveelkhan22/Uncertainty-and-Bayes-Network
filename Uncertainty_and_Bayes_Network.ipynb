```python
import pandas as pd
from collections import defaultdict

# Data from Table 2
data = [
    {'League': 'SerieA', 'Position': 'CF', 'Preferred Foot': 'Left', 'Capped': 'Yes', 'Shortlisted': True},
    {'League': 'LaLiga', 'Position': 'LW', 'Preferred Foot': 'Right', 'Capped': 'No', 'Shortlisted': True},
    {'League': 'LaLiga', 'Position': 'CF', 'Preferred Foot': 'Right', 'Capped': 'Yes', 'Shortlisted': False},
    {'League': 'Premier League', 'Position': 'CF', 'Preferred Foot': 'Left', 'Capped': 'Yes', 'Shortlisted': True},
    {'League': 'Premier League', 'Position': 'LW', 'Preferred Foot': 'Left', 'Capped': 'Yes', 'Shortlisted': False},
    {'League': 'Serie A', 'Position': 'RW', 'Preferred Foot': 'Left', 'Capped': 'Yes', 'Shortlisted': False},
    {'League': 'Serie A', 'Position': 'CF', 'Preferred Foot': 'Right', 'Capped': 'No', 'Shortlisted': True},
    {'League': 'Premier League', 'Position': 'LW', 'Preferred Foot': 'Left', 'Capped': 'No', 'Shortlisted': False},
    {'League': 'Serie A', 'Position': 'RW', 'Preferred Foot': 'Right', 'Capped': 'No', 'Shortlisted': True},
    {'League': 'LaLiga', 'Position': 'RW', 'Preferred Foot': 'Left', 'Capped': 'No', 'Shortlisted': False},
    {'League': 'LaLiga', 'Position': 'RW', 'Preferred Foot': 'Right', 'Capped': 'No', 'Shortlisted': True},
    {'League': 'Premier League', 'Position': 'CF', 'Preferred Foot': 'Right', 'Capped': 'No', 'Shortlisted': False},
    {'League': 'LaLiga', 'Position': 'CF', 'Preferred Foot': 'Right', 'Capped': 'Yes', 'Shortlisted': True},
    {'League': 'Serie A', 'Position': 'LW', 'Preferred Foot': 'Left', 'Capped': 'No', 'Shortlisted': False}
]

df = pd.DataFrame(data)

# Calculate prior probabilities
prior_prob = df['Shortlisted'].value_counts(normalize=True).to_dict()

# Function to calculate conditional probabilities for an attribute given class label
def conditional_probs(df, feature, class_label):
    counts = df[df['Shortlisted'] == class_label][feature].value_counts()
    total = counts.sum()
    return (counts / total).to_dict()

# Calculate conditional probabilities
conditional_probabilities = {'True': defaultdict(dict), 'False': defaultdict(dict)}
for label in [True, False]:
    for feature in ['League', 'Position', 'Preferred Foot', 'Capped']:
        conditional_probabilities[str(label)][feature] = conditional_probs(df, feature, label)

# Function to calculate Naive Bayes probability for a given player
def naive_bayes_predict(player, prior_prob, conditional_probabilities):
    probabilities = {}
    for label in [True, False]:
        prob = prior_prob[label]
        for feature, value in player.items():
            prob *= conditional_probabilities[str(label)][feature].get(value, 1e-6)  # Small value to handle zero counts
        probabilities[label] = prob
    # Normalize probabilities
    total = sum(probabilities.values())
    for label in probabilities:
        probabilities[label] /= total
    return probabilities

# Players for prediction
player1 = {'League': 'LaLiga', 'Position': 'RW', 'Preferred Foot': 'Right', 'Capped': 'Yes'}
player2 = {'League': 'Premier League', 'Position': 'LW', 'Preferred Foot': 'Left', 'Capped': 'No'}

# Predicting for each player
prediction1 = naive_bayes_predict(player1, prior_prob, conditional_probabilities)
prediction2 = naive_bayes_predict(player2, prior_prob, conditional_probabilities)

print("Player 1 Prediction:", prediction1)
print("Player 2 Prediction:", prediction2)

# Decision based on the highest probability for each player
decision1 = "Shortlisted" if prediction1[True] > prediction1[False] else "Not Shortlisted"
decision2 = "Shortlisted" if prediction2[True] > prediction2[False] else "Not Shortlisted"

print(f"Player 1 would be: {decision1}")
print(f"Player 2 would be: {decision2}")
```

```python
import pandas as pd
import numpy as np
import re
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# Load datasets (replace file paths with actual paths to your files)
newsgroup_train = pd.read_csv("newsgroup_train.csv")
newsgroup_test = pd.read_csv("newsgroup_test.csv")

# Extract features and labels
X_train = newsgroup_train['Text']  # Assuming 'Text' column holds the message content
y_train = newsgroup_train['Label']  # Assuming 'Label' column holds the category

X_test = newsgroup_test['Text']
y_test = newsgroup_test['Label']

class NaiveBayesTextClassifier:
    def __init__(self):
        self.word_counts = {}
        self.class_counts = {}
        self.vocab = set()
        self.total_counts = {}

    def preprocess(self, text):
        """Convert to lower case and remove non-word characters."""
        if isinstance(text, str):  # Check if the text is a string
            return re.findall(r'\b\w+\b', text.lower())
        else:
            return []  # Return empty list for non-string entries

    def fit(self, X, y):
        """Train the Naive Bayes classifier."""
        for text, label in zip(X, y):
            words = self.preprocess(text)
            self.vocab.update(words)  # Update vocabulary
            if label not in self.word_counts:
                self.word_counts[label] = {}
                self.class_counts[label] = 0
            self.class_counts[label] += 1

            for word in words:
                if word not in self.word_counts[label]:
                    self.word_counts[label][word] = 0
                self.word_counts[label][word] += 1

        # Total word counts for each class
        self.total_counts = {label: sum(counts.values()) for label, counts in self.word_counts.items()}

    def predict(self, X):
        """Predict class labels for the given text."""
        predictions = []
        vocab_size = len(self.vocab)

        for text in X:
            words = self.preprocess(text)
            class_scores = {}

            for label in self.class_counts.keys():
                prior = np.log(self.class_counts[label] / sum(self.class_counts.values()))
                likelihood = 0

                for word in words:
                    word_count = self.word_counts[label].get(word, 0)
                    # Apply Laplace smoothing
                    likelihood += np.log((word_count + 1) / (self.total_counts[label] + vocab_size))

                class_scores[label] = prior + likelihood

            # Choose the class with the highest score
            predictions.append(max(class_scores, key=class_scores.get))

        return predictions

# Initialize, train, and test the classifier
classifier = NaiveBayesTextClassifier()
classifier.fit(X_train, y_train)
predictions = classifier.predict(X_test)

# Evaluate
print("Accuracy:", accuracy_score(y_test, predictions))
print("Precision:", precision_score(y_test, predictions, average='macro'))
print("Recall:", recall_score(y_test, predictions, average='macro'))
print("F1 Score:", f1_score(y_test, predictions, average='macro'))
print("Confusion Matrix:\n", confusion_matrix(y_test, predictions))
```
